```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.svm import SVC
```

```python
def plot_moons_dataset(X, y):
    plt.figure(figsize=(6, 5))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
    plt.title("Raw Moons Dataset")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
def plot_svm_decision_boundary(model, X, y, title):
    # Create mesh grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 500),
        np.linspace(y_min, y_max, 500)
    )

    # Predict on grid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision regions
    plt.figure(figsize=(6, 5))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
X, y = make_moons(
    n_samples=300,
    noise=0.2,
    random_state=42
)
```

```python
plot_moons_dataset(X, y)
```

```python
linear_svm = SVC(kernel='linear')
linear_svm.fit(X, y)

plot_svm_decision_boundary(
    linear_svm,
    X,
    y,
    title="Linear SVM Decision Boundary"
)
```

```python
rbf_svm = SVC(kernel='rbf', gamma='scale')
rbf_svm.fit(X, y)

plot_svm_decision_boundary(
    rbf_svm,
    X,
    y,
    title="Non-Linear SVM (RBF Kernel) Decision Boundary"
)
```
